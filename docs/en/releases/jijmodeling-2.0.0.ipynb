# JijModeling 2.0.0 Release Notes

This release is the first major update since the JijModeling 1.x series.
While inheriting the spirit of JijModeling, it introduces various breaking improvements, such as support for comprehensions and the deprecation of the `Element` object.
For how to formulate and solve mathematical models with JijModeling 2, see **{doc}`JijModeling Basics<../basics/overview>`**. For details on changes and migration from JijModeling 1 to 2, see **{doc}`../references/migration_guide_to_jijmodeling2.md`**.
